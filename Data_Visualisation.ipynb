```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
bar_chart_data = pd.read_csv("bar_chart_homework_data.csv")
```

```python
bar_chart_data
```

```python
plt.figure(figsize = (9,6))
plt.bar(x = bar_chart_data["Cities"],
        height = bar_chart_data["Frequency"],
       color = "m")
plt.xticks(rotation = 30)
plt.yticks(rotation = 30)
plt.ylabel("Frequency", weight = "bold")
plt.xlabel("Cities", weight = "bold")
plt.title("Sales per City", fontsize = 18, weight = "bold")
plt.show()
```

```python
df_pie_chart = pd.read_csv("pie_chart_data.csv")
df_pie_chart
```

```python
plt.figure(figsize = (9, 9))
plt.pie(df_pie_chart["Number of Cars"],
       labels = df_pie_chart["Engine Fuel Type"],
       autopct = "%.2f%%",
       textprops = {"color" : "black",
                    "weight": "bold",
                   "size": "large"})
plt.title("Car Engine Types", fontsize = 20, weight = "bold")
plt.legend()
plt.show()
```

```python
df_area_chart = pd.read_csv("stacked_area_chart_data.csv")
```

```python
df_area_chart
```

```python
plt.figure(figsize = (12, 6))
plt.stackplot(df_area_chart["Year"],
               df_area_chart["Gas"],
               df_area_chart["Petrol"],
               df_area_chart["Diesel"],
             colors = ["navy", "cornflowerblue", "violet"],
             edgecolor = "none")
plt.xticks(rotation = 35)
plt.legend(labels = ["Gas", "Petrol", "Diesel"], loc = "upper left")
plt.title("Number of Car Engine Types 1982 - 2016", size = 18, weight = "bold")
plt.ylabel("Number of Cars")
plt.xlabel("Year")
```
